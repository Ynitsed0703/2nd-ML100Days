```python
import os
import pandas            as pd
import numpy             as np
import seaborn           as sns
import matplotlib.pyplot as plt
import math
import warnings
plt.style.use('ggplot')
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# 原本隨機 0.0 ~ 1.0, 減 0.5 乘兩倍 -> -1.0 ~ 1.0
# .random(y,x)
# y軸要12個，x軸要10個
# y軸由上至下從0~11
# x軸由左至右從0~ 9
# np.random是從0~1
matrix = (np.random.random((12,10)) - 0.5) * 2

# 圖片大小，要先寫
plt.figure(figsize=(10,10))

# 使用 seaborn heatmap
# 第一個參數：matrix，長相在上面了，matrix = (np.random.random((12,10)) - 0.5) * 2
# 第二個參數：plt.cm.xxxxxx是要用哪些色塊，Rd:red(紅),Yl:Yellow(黃),Bu:Blue(藍)，
#           不加【_r】是紅色從極小值開始，加【_r】是紅色從極大值開始
# RdYlBu這個沒辦法自己亂換，有哪些可以用，可參考此網站
# https://blog.csdn.net/guduruyu/article/details/60868501
# 第三個參數：設定色塊的極小值，例如若我在此範例設定【vmin =0】則小於0的會全部被設定成同個顏色，反之同理。
# 第四個參數：設定色塊的極大值，道理同上。
# 第五個參數：要不要秀出該色塊的數值？ True = 要 ； False = 不要
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r , vmin = -1., vmax = 1., annot = True)
plt.show()
```

```python
print('原始')
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1 , vmax = 1 , annot = True)
plt.show()

print('色塊參數沒加【_r】')
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu  , vmin = -1 , vmax = 1 , annot = True)
plt.show()

print('最小值設定成0')
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin =  0 , vmax = 1 , annot = True)
plt.show()

print('色塊改成Gn')
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlGn_r, vmin = -1 , vmax = 1 , annot = True)
plt.show()

print('annot = False')
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1 , vmax = 1 , annot = False)
plt.show()
```

```python
# 這個設計是建立出3000個位置喔。
ncol = 4
nrow = 1000

# 原本隨機 0.0 ~ 1.0, 減 0.5 乘兩倍 -> -1.0 ~ 1.0
matrix = (np.random.random((nrow,ncol)) - 0.5)*2
print(matrix)
```

# 說明1：Grid圖形與資料意義
1. Grid目前看來，是一個正方形矩陣，固定就是n*n的長相，而n是什麼？取決於你給的ncol是多少，再次證明昨天說的，要先決定ncol這件事。
2. 【ncol=3】的意義是什麼？表示給定三個欄位，三個欄位的命名分別是0,1,2，  
   現在打算觀察三個欄位的關係，每一筆資料，可以當作從3個欄位抽1筆資料，連續抽n次。  
   只是現在三個欄位我們都設計成一樣的資料狀態，要嘛uniform要嘛normal。  
3. 抽n次？到底是抽幾次，這就是【nrow=1000】，抽1000次就對了，換句話說它對grid劃出3*3這種長相並沒有影響，而是對於每一格裡面的長相有影響。  
4. 舉例最左下角的圖，列位名稱是2，欄位名稱是0，代表的意思是？  
   代表你拿【欄位0】和【欄位2】做交叉觀察，在欄位0出現1時，欄位2出現-3~+3的次數如何，  
   顯然這是一個3D圖，平面只看得出兩個normal交叉的分布狀況，卻看不出哪個地方會交叉得最多次，  
   這時候就需要顏色的輔助了，以上圖為例，顏色越深，表示出現的狀況越高。  
5. 舉例最右上角的圖，同理，只是現在改成了散點圖，也很明顯，越中間的地方，顯然有越多出現的機會。  
6. 對角線怎麼辦呢？因為對角線等於是【欄位n】和【欄位n】自己的比較，  
   而grid.map_diag(sns.kdeplot)這個函數很聰明，它知道這時候只要畫2D就可以了，就是簡單的機率分配。  

# 說明2：KDE的意義,[參考1](https://www.itread01.com/content/1546539491.html),[參考2](https://lotabout.me/2018/kernel-density-estimation/),[參考3](http://blog.shaochuancs.com/statistics-kde/)
所有平滑的峰值函數均可作為KDE的核函數來使用，常用的包括：矩形、Epanechnikov曲線、高斯曲線等。  

這東西沒有那麼困難，從直方圖、長條圖的角度開始出發即可。  
平常我們了解的直方圖、長條圖，每一個區間會有一根棒子，對吧？將很多區間的每根棒子合起來，就變成長條圖了嘛~  
所以我們的長條圖不是一個平滑的圖形，長條圖的上緣，是一堆連續的直角直角，而KDE可以很簡單的這樣想：  
就是把這一堆直角，畫成平滑的圖形，而一個平滑的圖形(線)，它肯定可以用連續型的函數去表示，就是這樣而已。  
至於這個連續型的函數要怎麼推導，那就是統計學課本裡面的東西了。
