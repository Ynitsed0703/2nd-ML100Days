```python
# 程式區塊 A：將需要的都import進來，設定好要使用的data路徑。
import os
import pandas            as pd
import numpy             as np
import seaborn           as sns
import matplotlib.pyplot as plt
import math
import warnings
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model  import LogisticRegression
warnings.filterwarnings('ignore')
%matplotlib inline

# 設定【data的資料夾路徑】，命名為【data_folder】
data_folder = 'C:/Users/Ynitsed/Documents/GitHub/2nd-ML100Days/data'
```

```python
# 設定t001為某個data路徑
# 設定t002為pd裡read data的功能
t001_train = os.path.join(data_folder, 'titanic_train.csv')
t002_train = pd.read_csv(t001_train)
print('Path of read in data: %s' %t001_train)
print(t002_train.shape)
t002_train.head()
```

```python
# 設定t001為某個data路徑
# 設定t002為pd裡read data的功能
t001_test  = os.path.join(data_folder,  'titanic_test.csv')
t002_test  = pd.read_csv(t001_test)
print('Path of read in data: %s' %t001_test)
print(t002_test.shape)
t002_test.head()
```

## 說明1
從train和test資料可發現，train有【Survived】但test沒有，可以想像後面是要讓test算出【Survived】的機率，只是透過什麼去算呢？

```python
# 程式區塊 B-1：做t003，train取【Survived】、test取【PassengerId】
t003_train = t002_train['Survived']
t003_train.head(10)
```

```python
t003_test  = t002_test['PassengerId']
t003_test.head(10)
```

```python
# 程式區塊 B-2：做t004，train捨棄【PassengerId,Survived】、test捨棄【PassengerId】。
t004_train = t002_train.drop(['PassengerId', 'Survived'] , axis=1)
t004_train.head(10)
```

```python
t004_test  = t002_test.drop(['PassengerId'] , axis=1)
t004_test.head(10)
```

```python
# 程式區塊 B-3：設計t005，把train和test合併，UNION ALL的概念。
# 現在還不太知道為什麼兩個都要把【PassengerId】拿掉，照理講train拿掉【Survived】就可以了。
t005 = pd.concat([t004_train,t004_test])
print(t005.shape)
t005.head(10)
```

```python
t005.describe()
```

```python
# 程式區塊 C：屬於類別欄位的用LE轉，屬於數值欄位的用MME轉。
LE = LabelEncoder()
MME = MinMaxScaler()
for d in t005.columns:
    t005[d] = t005[d].fillna(-1)
    if t005[d].dtype == 'object':
        t005[d] = LE.fit_transform(list(t005[d].values))
    t005[d] = MME.fit_transform(t005[d].values.reshape(-1, 1))
t005.head(10)
```

```python
t005.describe()
# 可以發現全部的欄位都被整理成0~1間的measure
```

```python
# 程式區塊 D-1：做一下train的總筆數
train_cnt = t002_train.shape[0]
train_cnt
```

```python
# 程式區塊 D-2：從頭列到某個段落，留意一下寫法[:cnt]；從中間列到尾，留意一下寫法[cnt:]
# 剛剛UNION ALL之後，做了一次正規化，現在要再把train和test拆開。
t006_train = t005[:train_cnt]
t006_train
```

```python
t006_test = t005[train_cnt:]
t006_test 
```

```python
# 程式區塊 D-3：拿目標欄位也就是被截取的【Survived】(t003_train)，去和被正規化後全部的欄位(t006_train)做【estimator.fit】
# 留意一下前後的關係，看起來是【estimator.fit(X,Y)】，Y是結果嘛。
estimator = LogisticRegression()
t007 = estimator.fit(t006_train, t003_train)
print(t007)
```

```python
# 程式區塊 D-4：t007可以當作已經跑完令人無法參透的運算，已經得到某個迴歸函數，現在我們要運用t007這個回歸函數，去套進test的裡。
# 換句話說，將t007這個回歸函數，實際填上test正規化後的各欄位數據(t006_test)，可以想像我們填了x1~xn進去。
# t006_test是x1~xn，t007則是b0和b1~bn，於是就能產生y了(t008)。
# 然而因為這是LogisticRegression，所以t008會再加一層，讓它轉成0(負數)和1(0&正數)。
t008 = t007.predict(t006_test)
t008
```

```python
# 程式區塊 E：將y的結果填回唯一識別欄位(不是用比對的(join))，看起來是按照順序填上。好吧~ 不會錯就好。
# pd.DataFrame(col_A,col_B)，而兩欄位的筆數必須相同。留意一下欄位命名的方式。
t009 = pd.DataFrame({'PassengerId': t003_test, 'Survived': t008})
t009.to_csv('titanic_baseline.csv', index=False) 
t009.head(10)
```

### Day17教材方向和目標

t003：train的部分是先擷取主要想被得知的欄位，test的部分則是擷取要被識別唯一的欄位。  
　　　更明確來說，我們最後想要知道，test資料中每一個【PassengerId】(唯一識別欄位)的【Survived】狀況，  
　　　因為test資料裡還沒有【Survived】，這就是我們要做的事。  
　　　所以t003，train的部分擷取【Survived】、test的部分擷取【PassengerId】，這是第一預備動作。  

t004：為了讓train和test的其他欄位被同時正規化，所以要把train和test併在一起，但兩張表的欄位不一樣啊，  
　　　train多出了【Survived】，所以要把它drop掉。  
　　　唯一識別欄位【PassengerId】drop掉的原因可能有兩個：

t005：處理完的t004終於可以UNION ALL了，在Python裡是【pd.concat([A_data,B_data])】。  
　　　接著把所有數據都正規化(0~1)，屬於類別欄位的用LE轉，屬於數值欄位的用MME轉。    

t006：既然正規化完了，那就把train和test再拆開來。這個拆法利用計算train的筆數去拆，挺不錯的。

t007：利用train的目標欄位(t003)和所有正規化欄位(t006)，運用LogisticRegression跑出回歸函數，b0,b1~bn，但沒辦法秀出來。
t008：將t006_test，也就是x1~xn套到回歸函數(t007)裡，可以得到每一個y1~yi
t009：t009將t008和最早取得的t003_test合併，可得最終結果。
