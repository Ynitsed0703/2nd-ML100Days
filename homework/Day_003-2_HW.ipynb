```python
## 假如我們不想把資料載到自己的電腦裡?
# 把連結填入
target_url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
```

```python
import requests
response = requests.get(target_url)
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[0:100]
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = '\n'

data = data.split(split_tag)
print(len(data))
data[0]
```

## 將 txt 轉成 pandas dataframe

```python
num = []
link = []
for d in data:
    if '\t' in d:
        d = d.split('\t')
        num.append(d[0])
        link.append(d[1])
    
labels = ['Num', 'Link']
cols = [num, link]

arrange_data = dict(list(zip(labels, cols)))
```

```python
import pandas as pd

    
df = pd.DataFrame(arrange_data)
df.head()
```

```python
def img2arr_fromURLs(url_list, resize = False):
    """
    請完成這個 Function
    Args
        - url_list: list of URLs
        - resize: bool
    Return
        - list of array
    """
    img_list = []
    for url in url_list:
        img = None
        try:
            response = requests.get(url)
            img = Image.open(BytesIO(response.content))
        except:
            print('No response: ', url)
    
        if img is not None:
            img_list.append(img)
            
    return img_list    
    
    
    
    return img_list
```
