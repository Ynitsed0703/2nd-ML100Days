```python
# 載入基礎套件
# 不懂【np.random.seed(1)】的意義用意
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 弱相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何
# Day9真的換過老師了啦~ 這個講得有夠清楚的。Day8之前參數在幹嘛都不解釋的。

x_sample1 = np.random.randint(0, 50, 1000)
y_sample1 = np.random.randint(0, 50, 1000)

# 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x_sample1, y_sample1)

#跑出來的結果不知道是什麼意思...為什麼不解釋？
#左上角【1.】應該是【x對x】
#第二個和第三個【0.00404702】，應該分別是【x對y】和【y對x】
#右下角【1.】應該是【y對y】
```

```python
# 將分布畫出來看看吧
plt.scatter(x_sample1, y_sample1)
```

### 正相關

```python
# 隨機生成 1000 個介於 0~50 的數 x
x_sample2 = np.random.randint(0, 50, 1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
# 又不解釋了，欸這真的是齁？為什麼不多說幾句呢？
# 雜訊：0~10的常態分佈，而且不限定是整數。
y_sample2 = x_sample2 + np.random.normal(0, 10, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x_sample2, y_sample2)
```

```python
# 再看看正相關的 x,y 分布
plt.scatter(x_sample2, y_sample2)
```

# 作業
- 參考範例程式碼，模擬一組負相關的資料，並計算出相關係數以及畫出 scatter plot

```python
# 隨機生成 1000 個介於 0~100 的數 x
x_hw1 = np.random.randint(0, 100, 1000)

# 這次讓 y 與 x 負相關，再增加一些雜訊 (0~5)
y_hw1 = - x_hw1 + np.random.normal(0, 5 , 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x_hw1, y_hw1)
```

```python
# 隨機生成 1000 個介於 0~100 的數 x
x_hw2 = np.random.randint(0, 100, 1000)

# 這次讓 y 與 x 負相關，再增加一些雜訊 (0~50)
# 雜訊變大，預期相關係數要變小(變更無關)
y_hw2 = - x_hw2 + np.random.normal(0, 50, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x_hw2, y_hw2)
```

```python
# 再看看負相關的 x,y 分布
plt.scatter(x_hw1, y_hw1)
```

```python
# 再看看負相關的 x,y 分布
plt.scatter(x_hw2, y_hw2)
```
